Confidence Interval

```python
sample_1 = sample_df['Medical Care Expenditure']
```

```python
sample_1_mean = sample_1.mean()
```

```python
t_critical_1 = stats.t.ppf(q = 0.975, df=24)
```

```python
print("t-critical value:")
print(t_critical_1)
```

```python
sample_1_stdev = sample_1.std(ddof=1)
```

```python
sample_size = 100
```

```python
sigma_1 = sample_1_stdev/math.sqrt(sample_size)
margin_of_error_1 = t_critical_1 * sigma_1
```

```python
confidence_interval = (sample_1_mean - margin_of_error_1,
                       sample_1_mean + margin_of_error_1) 
```

```python
print("Confidence interval_1:")
print(confidence_interval)
```

```python
sample_2 = sample_df['Education Expenditure']
```

```python
sample_2_mean = sample_2.mean()
```

```python
t_critical_2 = stats.t.ppf(q = 0.975, df=24)
```

```python
print("t-critical value:")
print(t_critical_2)
```

```python
sample_2_stdev = sample_2.std(ddof=1)
```

```python
sample_size = 100
```

```python
sigma_2 = sample_2_stdev/math.sqrt(sample_size)
margin_of_error_2 = t_critical_2 * sigma_2
```

```python
confidence_interval = (sample_2_mean - margin_of_error_2,
                       sample_2_mean + margin_of_error_2)
```

```python
print("Confidence interval_2:")
print(confidence_interval)
```

**Calcule se há diferença de média de número total de membros da família pelo sexo do chefe de família (household head sex e total number of Family members) com 95% de confiança.**

```python
sample_df_male = sample_df[sample_df['Household Head Sex'] == "Male"]
```

```python
mean_male = sample_df_male['Total Number of Family members'].mean()
```

```python
sample_df_female = sample_df[sample_df['Household Head Sex'] == "Female"]
```

```python
mean_female = sample_df_female['Total Number of Family members'].mean()
```

```python
stats.ttest_ind(sample_df_male['Total Number of Family members'],
                sample_df_female['Total Number of Family members'],
                equal_var=False)  
```

De acordo com o Pvalue H0 é aceita, logo pode se considerar que há diferença entre as médias.

**Calcule se há diferença de média total de renda da família(total household Income) de acordo com a fonte salarial(main source of income) Tarefa: considere “Enteroreneurial activities” e “other sources of income” como: outros.**

```python
sample_df['Total Household Income']
```

```python
sample_df['Main Source of Income']
```

```python
def tarefa(x):
    if x == 'Enterpreneurial Activities' or x == 'Other sources of Income':
        return 'outros'
    else:
        return x
```

```python
sample_df['fonte_renda'] = sample_df['Main Source of Income'].apply(tarefa)
```

```python
sample_df['fonte_renda'].value_counts()
```

```python
sample_df_outros = sample_df[sample_df['fonte_renda'] == "outros"]
```

```python
sample_df_Salaries = sample_df[sample_df['fonte_renda'] == "Wage/Salaries"]
```

```python
stats.ttest_ind(sample_df_outros['Total Household Income'],
                sample_df_Salaries['Total Household Income'],
                equal_var=False)  
```

De acordo com o Pvalue H0 é aceita, logo pode se considerar que há diferença entre as médias.
